#### Function to get the names out of Dictionary of lists

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

This function extracts only the name pair from the dictionary.

What it does:
Takes a string that looks like a list of dicts.

Extracts and returns only the "name" values.



```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

Now Applying the same function in Keywords,cast and Crew columns

```python
movies["keywords"]= movies["keywords"].apply(convert)
```

#### Functio to Limit Genre Extraction to First 3 Names

```python
movies["cast"][0]
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

 What This Does:
ast.literal_eval(obj): Safely converts the string into a real Python list of dictionaries.

- L = []: Empty list to store genre names.
- counter = 0: Keeps track of how many names have been added.
- if counter != 3: Add names only up to 3.
- break: Stop the loop after getting 3 names.
- return L: Gives back a list of the first 3 names.

```python
movies["cast"]= movies["cast"].apply(convert3)
```

```python
movies.head()
```

#### Function to Fetch only director out of other Crew members

```python
movies["crew"].iloc[0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

 What This Does:
- ast.literal_eval(obj): Converts the string into a list of dictionaries (crew data).
- Checks each item in the list:

    - If "job" is "Director", adds "name" to the list L.

    - break: Stops after finding the first director.

- Returns the list L containing the director’s name.

```python
movies["crew"]= movies["crew"] .apply(fetch_director)
```

## Text Vectorization

#### Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

What is PorterStemmer? 
PorterStemmer also known as Stemming reduces words to their root form.

Example:

- "running" → "run"

- "loving" → "love"
- "liked" → "like"

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

stem(text) Function Breakdown:
- Splits the input text into individual words using .split().

- Applies PorterStemmer to each word to get its root form.
- Stores the stemmed words in a list y.

- Joins the stemmed words back into a single string.

- Returns the final stemmed sentence.

```python
df["tags"]= df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features= 5000, stop_words="english")
```

```python
vectors = cv.fit_transform(df["tags"]).toarray()
```

#### Cosine Similarity

💡 What it does:
This line imports the cosine_similarity function, which is used to measure how similar two vectors are.

✅ Use Case:
It’s commonly used in:

- Text similarity

- Recommendation systems (e.g., finding similar movies)

📌 Output:
- It gives a similarity matrix — higher values mean more similar. 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

## Main Function

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
df[df["title"]=="Batman Begins"].index[0]
```

```python
def recommend(movie):
# ✅ Step 1: Get index of the movie
    movie_index = df[df["title"] == movie].index[0]
# 📊 Step 2: Get similarity scores for this movie
    distances = similarity[movie_index]
 # 📌 Step 3: Sort all movies by similarity score (highest first)   
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

# 🖨️ Step 4: Print top 5 similar movie titles
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

🧱 Bullet Point Summary:
- 🔹 movie_index = ... → Finds the row number of the input movie.

- 🔹 distances = ... → Gets similarity scores of all movies with this one.

- 🔹 sorted(...)[1:6] → Sorts and selects top 5 (skips the movie itself).

- 🔹 print(...) → Prints titles of similar movies.

```python
recommend("Superman")
```
